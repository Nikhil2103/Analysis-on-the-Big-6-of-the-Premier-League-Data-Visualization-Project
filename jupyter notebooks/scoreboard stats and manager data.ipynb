```python
df_managers["Win_Percentage"]= round(df_managers["Won"]/df_managers["Played"] * 100,2)
df_managers["Loss_Percentage"]= round(df_managers["Lost"]/df_managers["Played"] * 100,2)
```

```python
final_managers = df_managers.groupby("Managers").mean("Win_Percentage")[["Win_Percentage","Loss_Percentage"]]
```

```python
matches_played = df_managers.groupby("Managers").sum("Played")[["Played"]]
```

```python
df_managers.groupby("Managers").mean("Win_Percentage")
```

```python
top10_managers_winloss_percentage = final_managers.sort_values("Win_Percentage", ascending=False)[:10]
```

#### getting the top 10 managers stats

```python
top10_managers_winloss_percentage
```

```python
matches_played.index.values
```

```python
matches_played.reset_index(inplace = True)
```

```python
matches_played
```

```python
played_games =  matches_played[matches_played["Managers"].isin(top10_managers_winloss_percentage.index.values)]
```

```python
played_games.set_index("Managers",inplace = True)
```

```python
top10_managers_winloss_percentage = top10_managers_winloss_percentage.join(played_games)
```
